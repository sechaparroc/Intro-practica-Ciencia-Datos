# Breve introducción a NumPy - Representando  y preprocesando los datos

En Python existe una estructura de datos que se conoce como lista y permite almacenar distintos tipos de datos.

NumPy (Numeric Python) permite realizar distintas operaciones útiles en el procesamiento de datos sobre listas con información numérica (arreglos multidimensionales) de manera rápida e intuitiva.

Por ejemplo, supongamos que tenemos la información de la altura de diversas personas (en pulgadas) y su peso (en libras).

¿Cómo podríamos resolver los siguientes problemas?

1. Pasar la estatura de pulgadas a centimetros
2. Pasar el peso de libras a kilogramos
3. Calcular el indice de masa corporal

Con Numpy esto es muy sencillo, pues podemos aplicar operaciones aritméticas sobre todo el conjunto de datos!
Para crear un arreglo de Numpy, podemos utilizar la funcion ``np.array()`` y como parámetro utilizar una lista que queremos convertir.

Tenga en cuenta que los arreglos de Numpy **solo son de un tipo de dato**.


````python
import numpy as np

altura_en_pulgadas = np.array([73.85, 68.78, 74.11, 71.73, 69.88, 67.25, 68.79, 68.35, 67.02, 63.46, 71.2, 71.64, 64.77, 69.28, 69.24, 67.65, 72.42, 63.97, 69.64, 67.94, 67.92, 69.44, 66.15, 75.21, 67.89, 68.14, 69.09, 72.8, 67.42, 68.5, 68.62, 74.03, 71.53, 69.18, 69.58, 70.4, 69.08, 67.19, 65.81, 64.3, 67.97, 72.19, 65.27, 66.09, 67.51, 70.1, 68.25, 72.17, 69.18, 72.87])
peso_libras = np.array([241.89, 162.31, 212.74, 220.04, 206.35, 152.21, 183.93, 167.97, 175.93, 156.4, 186.6, 213.74, 167.13, 189.45, 186.43, 172.19, 196.03, 172.88, 185.98, 182.43, 174.12, 197.73, 149.17, 228.76, 162.01, 192.34, 184.44, 206.83, 175.21, 154.34, 187.51, 212.91, 195.03, 205.18, 204.16, 192.9, 197.49, 183.81, 163.85, 163.11, 172.14, 194.05, 168.62, 161.19, 164.66, 188.92, 187.06, 209.07, 192.01, 211.34])

altura_metros = altura_en_pulgadas * 0.0254
peso_kg = peso_libras * 0.453592

print(altura_metros)
print(peso_kg)
#altura y peso de la ultima persona
print("Datos de la ultima persona ", altura_metros[-1], peso_kg[-1])
````

## Conociendo nuestros datos
Queremos determinar a que tipo de personas encuestamos, una vez conocida su altura y peso usando el sistema internacional SI, es posible determinar el Índice de masa corporal de las personas. Esto nos puede dar indicios sobre el bienestar de las mismas.

````python
indice_masa_corporal = peso_kg / altura_metros ** 2
print(indice_masa_corporal)
#IMC de la ultima persona
print("IMC de la ultima persona", indice_masa_corporal[-1])
````

No solo se pueden realizar operaciones aritméticas, tambien podemos realizar comparaciones:

````python
indice_masa_corporal > 30
````

Mejor aún, podemos utilizar el arreglo booleano resultante para filtrar a aquellas personas con un indice superior a 30:

````python
indice_masa_corporal[indice_masa_corporal > 30]
````

````python
#con esta información podemos categorizar a las personas por su indice de masa corporal
print("personas con delgadez", len(indice_masa_corporal[indice_masa_corporal < 18]))
print("personas con peso adecuado", len(indice_masa_corporal[(18 <= indice_masa_corporal) & (indice_masa_corporal < 25)]))
print("personas con sobrepeso", len(indice_masa_corporal[(25 <= indice_masa_corporal) & (indice_masa_corporal <= 30)]))
print("personas con obesidad", len(indice_masa_corporal[indice_masa_corporal > 30]))
````

De la informacion anterior podemos observar que la mayoría de las personas se encuentran en sobrepeso.
Note que esta categorización es **explícita** por tanto no cuenta como un modelo de aprendizaje de máquina, sin embargo es útil para caracterizar a la población.

## Funciones lineplot y scatterplot

La función `sns.lineplot(x = ___ , y = ___)` Realiza una gráfica a partir de una serie de datos que representan el eje horizonal vs otros que representan el eje vertical. Cada dato puntual es unido al anterior mediante un segmento de línea. Este tipo de gráfica sirve para representar observaciones que siguen un orden.

La función `sns.scatterplot(x = ___ , y = ___)` Realiza una gráfica de dispersion (no une las observaciones) a partir de una serie de datos que representan el eje horizonal vs otros que representan el eje vertical.

Tenga en cuenta que las gráficas solo se muestran cuando se realiza la instrucción `plt.show()`.

````python
#modulos de python para realizar visualizaciones
import seaborn as sns 
import matplotlib.pyplot as plt

sns.scatterplot(x = peso, y = altura)
plt.show() # lo mostramos 
````

## Breve introducción a Sklearn - Dividiendo nuestros datos y seleccionando un modelo

Aunque podemos manipular los datos para obtener información nueva de ellos utilizando nuestra experticia, el aprendizaje de máquina nos permite delegar esta responsabilidad al computador, de modo que sea este quien facilite la toma de decisiones.

En nuestro ejemplo, podemos "enseñar" al computador a predecir cual es el sexo de una persona conocida su estatura y su peso.

Dado que conocemos las caracteristicas de entrada (peso, altura), la etiqueta de salida o variable dependiente (sexo) y tenemos datos previamente etiquetados podemos utilizarlos para entrenar un modelo que nos permita predecir la etiqueta de salida sobre nuevos datos.

Para realizar esta labor utilizaremos la librería **Sci-kit Learn**.
Note que nuestro problema se resume a lo siguiente: tenemos información **etiquetada** (datos de entrenamiento) y deseamos utilizarla para predecir la **etiqueta** de una observación nueva. ¿Cómo podemos construir un modelo de ML a partir de esta información?

### Clasificando con KNN (K - Nearest Neighbors) - El modelo de vecinos más cercanos 

La idea de este modelo consiste en predecir el valor de la observación nueva verificando cuales son las K observaciones etiquetadas más cercanas a esta.

Luego, determinaremos a que categoría pertenecerá la observación por votación (la etiqueta más frecuente entre los vecinos).

Coloquialmente podríamos decir que este modelo sigue el principio "Dime con quien andas y te diré quien eres".

Aunque existen muchos otros modelos, para el propósito de este Notebook introductorio solo se revisará KNN .

Todos los modelos de sci-kit learn son implementados como clases de Python y cuentan con las funciones `fit.()` para entrenamiento y `predict.()` para predicción.

[Click aquí para ver ejemplo](https://editor.p5js.org/sechaparroc/sketches/dzH3iWzoT)

````python
from sklearn.neighbors import KNeighborsClassifier #Importamos el modelo de clasificacion KNN
knn = KNeighborsClassifier(n_neighbors = 3) #creamos un nuevo modelo KNN que tendra en cuenta 3 vecinos

datos = np.array([peso, altura]).transpose() # hacemos que cada fila sea una observacion
etiquetas = sexo.transpose() # convertimos el arreglo en una columna

#podemos entrenar el modelo indicando cuales son las caracteristica y cuales las etiquetas esperadas
knn.fit(datos, etiquetas)
````

````python
#para predecir la etiqueta de un nuevo dato podemos pasar una tabla con diversas observaciones por fila
datos_sin_etiqueta = np.array([[50, 1.60],
                               [60, 1.80],
                               [70, 1.70],
                               [80, 1.90],])
#predecimos el valor de las etiquetas utilizando el modelo
etiquetas_predichas = knn.predict(datos_sin_etiqueta)
etiquetas_predichas
````

````python
#podemos graficar las predicciones en una grafica de dispersion
plt.figure(figsize=(15, 5))
sns.scatterplot(x = peso, y = altura, hue = sexo, alpha = 0.6)
sns.scatterplot(x = datos_sin_etiqueta[:,0], y = datos_sin_etiqueta[:,1], hue = etiquetas_predichas, s= 300, marker = "*")
plt.show() # lo mostramos 
````

````python
#probemos ahora con 5 vecinos... 
knn = KNeighborsClassifier(n_neighbors = 5) #creamos un nuevo modelo KNN que tendra en cuenta 3 vecinos
knn.fit(datos, etiquetas)
#predecimos el valor de las etiquetas utilizando el modelo
etiquetas_predichas = knn.predict(datos_sin_etiqueta)
etiquetas_predichas
````

````python
#podemos graficar las predicciones en una grafica de dispersion
plt.figure(figsize=(15, 5))
sns.scatterplot(x = peso, y = altura, hue = sexo, alpha = 0.6)
sns.scatterplot(x = datos_sin_etiqueta[:,0], y = datos_sin_etiqueta[:,1], hue = etiquetas_predichas, s= 300, marker = "*")
plt.show() # lo mostramos 
````

## Midiendo el desempeño del modelo

Hemos generado dos modelos con distintos parámetros pero ¿como podemos comporbar cual es el mejor de ellos?


Para medir el desempeño de los modelos de clasificación es común utilizar la **precisión** (accuracy), definida como la proporción de predicciones correctas.

Pero ¿qué datos podemos utilizar para medir la precisión?
Dado que no sabemos con certeza la etiqueta de los datos a predecir, lo único que podemos hacer es utilizar nuestros datos etiquetados. 

Para ello lo que hacermos será dividir nuestros datos en dos conjuntos: un conjunto de entrenamiento (train set), con el que entrenamos el modelo y un conjunto de pruebas (test set) que no se utilizarán en la fase de entrenamiento y en el que haremos predicciones para determinar la precisión del modelo.

````python
#sklearn permite dividir nuestros datos de forma sencilla
from sklearn.model_selection import train_test_split

#la funcion train_test_split permite dividir nuestros datos de forma aleatoria e indicando la proporcion de datos de pruebas
datos_entrenamiento, datos_prueba, etiquetas_entrenamiento, etiquetas_prueba = train_test_split(datos, etiquetas, test_size = 0.3, random_state = 0, stratify = etiquetas)
````

````python
#entrenamos el modelo de KNN
knn = KNeighborsClassifier(n_neighbors = 3) #creamos un nuevo modelo KNN que tendra en cuenta 3 vecinos
#usamos los datos de entrenamiento
knn.fit(datos_entrenamiento, etiquetas_entrenamiento)

#verificamos la precision del modelo con los datos de prueba
precision_3 = knn.score(datos_prueba, etiquetas_prueba)
print("La precisión es", precision_3)
````

````python
#entrenamos el modelo de KNN
knn = KNeighborsClassifier(n_neighbors = 5) #creamos un nuevo modelo KNN que tendra en cuenta 3 vecinos
#usamos los datos de entrenamiento
knn.fit(datos_entrenamiento, etiquetas_entrenamiento)

#verificamos la precision del modelo con los datos de prueba
precision_5 = knn.score(datos_prueba, etiquetas_prueba)
print("La precisión es", precision_5)
````

````python
# Verifiquemos la precision para distintos valores de k
precision = []
for i, k in enumerate(range(1,10)):
    # Setup a k-NN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors = k)
    # entrenamos el modelo
    knn.fit(datos_entrenamiento, etiquetas_entrenamiento)
    # calculamos la precision con los datos de prueba
    precision.append(knn.score(datos_prueba, etiquetas_prueba))

# Mostramos la grafica de vecinos vs precisión
plt.figure(figsize=(15, 5))
plt.title('Cantidad de vecinos')
plt.plot(range(1,10), precision, label = 'Precisión')
plt.legend()
plt.xlabel('Vecinos')
plt.ylabel('Precisión')
plt.show()
````

### Regesión con KNN (K - Nearest Neighbors) - El modelo de vecinos más cercanos 

Recuerde que en los modelos de regresión la variable dependiente o de salida es numérica y es continua.

Podemos usar KNN y predecir el valor de una observación verificando los valores de las K observaciones etiquetadas más cercanas a esta.

Para determinar el valor de la variable de salida, simplemente tomamos el promedio de los valores de sus vecinos.

Todos los modelos de sci-kit learn son implementados como clases de Python y cuentan con las funciones `fit.()` para entrenamiento y `predict.()` para predicción.

Para este ejemplo de regresión, trataremos de predecir el peso de una persona de acuerdo a su estatura y su sexo.

[Click aquí para ver un ejemplo gráfico](https://editor.p5js.org/sechaparroc/sketches/bHh06gK70)

````python
#Primero manipulemos los datos para que sean columnas (de modo que cada observacion sea una fila)
sexo = sexo.reshape(-1, 1)
altura = altura.reshape(-1, 1)

for i in range(0,10):
    print("Persona ", i, altura[i] , sexo[i], peso[i])
````

Es importante notar, que este modelo de ML requiere que las características de las observaciones sean numéricas. Así, que en el caso de tratar con variables categóricas es necesario **preprocesarlas**. Esto se puede hacer codificándolas mediante **"one hot encoders"** que generarán una lista de numeros con tantas posiciones como categorías en donde una posición será:
- 0 Si la observación no está dentro de la categoría 
- 1 Si la observación está dentro de la categoría 

````python
# Para preprocesar usamos la librería sklearn
from sklearn.preprocessing import OneHotEncoder # importamos el modulo de preprocesamiento
codificador = OneHotEncoder() #creamos el codificador
sexo_numerico = codificador.fit_transform(sexo).toarray() #aplicamos el codificador en el arreglo correspondiente al sexo de las personas
sexo_numerico[: 8]
````

````python
#Ahora dado que nuestras características ya son todas numericas las podemos representar en una unica tabla
datos = np.column_stack((altura , sexo_numerico))
print("Persona".center(10), "Altura".center(10), "Es Mujer".center(10), "Es Hombre".center(10), "Peso (Etiqueta)". center(15))
for i in range(0,10):
    print(str(i).center(10), str(round(datos[i,0],2)).center(10) , str(round(datos[i,1],2)).center(10), str(round(datos[i,2],2)).center(10), str(round(peso[i],2)).center(15))
````

Una vez todas nuestras características tienen un valor numerico podemos aplicar el modelo de regresión KNN. Aquí es **importante resaltar que es habitual y necesario [normalizar las características](https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/)** (en este ejemplo no se realiza porque las 3 primeras características tienen una escala similar).

````python
from sklearn.neighbors import KNeighborsRegressor #Importamos el modelo de regresion KNN
knn_regresion= KNeighborsRegressor(n_neighbors = 3) #creamos un nuevo modelo KNN que tendra en cuenta 3 vecinos
#podemos entrenar el modelo indicando cuales son las caracteristica y cuales las etiquetas esperadas
knn_regresion.fit(datos, peso) #la etiqueta en este caso es el peso
````

````python
#para predecir la etiqueta de un nuevo dato podemos pasar una tabla con diversas observaciones por fila
datos_sin_etiqueta = np.array([[1.65, 1, 0],
                               [1.65, 0, 1],
                               [1.75, 1, 0],
                               [1.75, 0, 1],])
#predecimos el valor de las etiquetas utilizando el modelo
etiquetas_predichas = knn_regresion.predict(datos_sin_etiqueta)
etiquetas_predichas
````

````python
# Podemos medir el desempeño para varios valores de K
#Dividimos nuestros datos
from sklearn.model_selection import train_test_split #sklearn permite dividir nuestros datos de forma sencilla
#la funcion train_test_split permite dividir nuestros datos de forma aleatoria e indicando la proporcion de datos de pruebas
datos_entrenamiento, datos_prueba, etiquetas_entrenamiento, etiquetas_prueba = train_test_split(datos, peso, test_size = 0.3, random_state = 0, stratify = etiquetas)

print(etiquetas_prueba)

#Probamos con distintos parametros
precision = []
for i, k in enumerate(range(1,10)):
    knn_regresion = KNeighborsRegressor(n_neighbors = k)
    # entrenamos el modelo
    knn_regresion.fit(datos_entrenamiento, etiquetas_entrenamiento)
    # calculamos la precision con los datos de prueba
    precision.append(knn_regresion.score(datos_prueba, etiquetas_prueba))

# Mostramos la grafica de vecinos vs precisión
plt.figure(figsize=(15, 5))
plt.title('Cantidad de vecinos')
plt.plot(range(1,10), precision, label = 'Precisión')
plt.legend()
plt.xlabel('Vecinos')
plt.ylabel('r^2')
plt.show()
````

## 7. Agregando nuevas columnas a nuestros DataFrames

Para agregar una nueva columna basta con realizar la siguiente operación: `tabla[nueva_col] = ________` donde en la parte izquierda realizaremos alguna operacion con arreglos.

## 9. Operaciones de Agregación

Para interpretar nuestros datos debemos realizar operaciones que nos permitan conocer características sobre ellos, habitualmente utilizamos estadisticos que corresponden a valores numéricos que "nos permiten comunicar la mayor cantidad de información de la manera más simple posible". Los siguientes corresponden a los estadísticos descriptivos más usados:

- `tabla[columna].mean()`    : Para determinar el promedio.
- `tabla[columna].mode()`    : Para determinar la moda.
- `tabla[columna].median()`  : Para determinar la mediana.
- `tabla[columna].min()`     : Para determinar el mínimo.
- `tabla[columna].max()`     : Para determinar el máximo.
- `tabla[columna].std()`     : Para determinar la desviación estandar.
- `tabla[columna].var()`     : Para determinar la varianza.
- `tabla[columna].quantile()`: Para determinar cuantiles.
- `.cumsum()`, `.cummin()`, `cummax()` :  Para determinar la suma, el mínimo y el máximo acumulado.


Los estadísticos anteriores solo pueden aplicarse a columnas que tengan valores numéricos. Para valores categóricos se puede utilizar la función `value_counts` que nos permitirá conocer cuantos datos hay de cada categoría.

Finalmente, pueden aplicarse estadísticos descriptivos agrupando los datos de acuerdo a las categorpias de una columna, para ello se utiliza la función `tabla.groupby("columna")` seguida de la aplicación de un estadístico descriptivo. Por ejemplo, podemos aplicar la función `.agg([max, min, mean])` para generar varios estadísticos al tiempo.


````python
# Podemos determinar cuales son las categorias de las species
tabla_aire_covid["Specie"].unique()
````

````python
tabla_aire_covid["Specie"].value_counts()
````

````python
#Podemos agrupar por tipo de contaminante y generar los estadisticos descriptivos del valor de la mediana de los contaminantes
tabla_aire_covid.groupby("Specie")["Value"].agg([np.max, np.min, np.mean, np.median])
````

````python
tabla_aire_covid_or = tabla_aire_covid.copy() #mantenemos la tabla original (si se llegase a requerir)
tabla_aire_covid = tabla_aire_covid.drop('count', 1) #Por simplicidad solo utilizaremos el valor de las "species"
````

````python
#Agrupemos por país tomando el promedio de las mediciones de las ciudades
tabla_aire_pais = pd.DataFrame()
for i,specie in enumerate(["temperature", "humidity", "pressure", "pm25", "pm10", "co"]):
    es_specie = tabla_aire_covid["Specie"] == specie
    if i == 0:
        tabla_aire_pais = tabla_aire_covid[es_specie].groupby(["Date","Country"]).mean()
        tabla_aire_pais = tabla_aire_pais.rename(columns = {"Value" : specie})
        
    else:
        aux = tabla_aire_covid[es_specie].groupby(["Date","Country"]).mean()
        aux = aux.rename(columns = {"Value" : specie})
        # pueden cambiar how = "outer" por how = "inner" y no se tendrán en cuenta los datos incompletos
        # si hacen eso el número de registros disminuye notablemente!
        tabla_aire_pais = tabla_aire_pais.merge(aux, on = ["Date", "Country"], how = "outer") 
        
tabla_aire_pais = tabla_aire_pais.reset_index(level='Country')
tabla_aire_pais
````

````python
# Creemos una nueva columna con los nombres reales
# A que código corresponden dichos paises ?
# Podemos utilizar otro archivo CSV para identificarlo!
codigo_pais = pd.read_csv('datasets/country_codes.csv')
codigo_pais = codigo_pais.set_index('Code')
codigo_pais.head()
````

````python
# Reemplacemos el código con el nombre del país 
tabla_aire_pais["Country"] = tabla_aire_pais["Country"].replace(to_replace = codigo_pais.index, value = codigo_pais["Name"] )
tabla_aire_pais
````
